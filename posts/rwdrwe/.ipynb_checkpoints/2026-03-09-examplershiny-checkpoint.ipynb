1. **데이터 구성**

- 데이터 출처: Electronic Data Capture (EDC) 시스템 기반 실시간 임상 데이터(임의 생성)
- 주요 데이터
    - 피험자 등록 현황 (Enrollment Status)
    - 이상반응 보고 (AE / SAE)
    - Query 발생 및 해결 현황
    - Site Monitoring Visit 일정 및 진행 상태
- 데이터 특징
    - 다기관 임상시험 데이터로 site 및 visit 기준의 다차원 구조
    - 지속적으로 업데이트되는 실시간 데이터
    - 분석 전 데이터 통합, 정제 및 집계 과정 수행

---

```r

# set seed 고정 후
sites <- paste0("Site_", sprintf("%03d", 1:10))
dates <- seq.Date(from = as.Date("2024-01-01"), to = as.Date("2024-12-31"), by = "day")

# 데이터 생성
df_list <- lapply(sites, function(s) {
  n <- length(dates)
  data.frame(
    site = s,
    date = dates,
    visit = sample(1:5, n, replace = TRUE),
    subjid = paste0(substr(s, 6, 8), sprintf("%04d", sample(1:300, n, replace = TRUE))),
    enrolled = sample(0:3, n, replace = TRUE),
    ae_count = sample(0:3, n, replace = TRUE),
    gender = sample(c("Male", "Female"), n, replace = TRUE)
  )
})

```

2. **분석 방법 및 기술**

- 사용 기술
    - R (dplyr, ggplot2, lme4)
    - R Shiny (shinyWidgets, shinythemes)
    - R Markdown
- 분석 방법
    - Mixed Effects Model
        - 반복 측정(longitudinal) 임상 데이터를 분석하여 피험자 내 상관 구조를 고려한 통계 분석 수행
            - 반복측정 데이터를 대상으로 수행한 선형 혼합모형의 고정효과 추정 결과를 보여줌,
            - visit는 방문 시점에 따른 점수 변화의 평균적인 경향을 평가하며, subject에 대한 랜덤 절편(random intercept)을 포함하여 개인 간 차이를 반영함.
    - 자동화 및 리포팅
        - KPI 요약 테이블 및 그래프를 자동 생성
        - R Markdown 기반 자동 보고서 생성 및 다운로드 기능 제공
