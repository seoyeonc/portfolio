## Import

```python
import random
import pandas as pd
```

```python
def block_randomization(n_subjects, block_size=4, treatments=['Test', 'Reference'], seed=123):
    random.seed(seed)
    
    if block_size % len(treatments) != 0:
        raise ValueError("block_size must be divisible by the number of treatments.")
    
    assignments = []
    n_per_treatment = block_size // len(treatments)
    
    n_blocks = (n_subjects + block_size - 1) // block_size
    
    for _ in range(n_blocks):
        block = []
        for trt in treatments:
            block.extend([trt] * n_per_treatment)
        random.shuffle(block)
        assignments.extend(block)
    
    assignments = assignments[:n_subjects]
    
    df = pd.DataFrame({
        'Subject_ID': [f'SUBJID_{i+1:03d}' for i in range(n_subjects)],
        'Treatment': assignments
    })
    
    return df
```

```python
df_rand = block_randomization(n_subjects=20, block_size=4, treatments=['Test', 'Reference'], seed=2026)
print(df_rand)
```

```python
print("\nTreatment counts:")
print(df_rand['Treatment'].value_counts())
```
